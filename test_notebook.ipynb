```python
# Now create a chart using matplotlib to show the token balances in a pie chart.

def create_chart(api_response):
    # create a list of token balances
    balances = []
    for token in api_response['balances']:
        balances.append(token['balanceUSD'])

    # create a list of token symbols
    symbols = []
    for token in api_response['balances']:
        symbols.append(token['symbol'])

    # create a pie chart
    import matplotlib.pyplot as plt
    plt.pie(balances, labels=symbols)
    plt.show()

create_chart(api_response=response.json())
```

```python
# Now create a chart using matplotlib to show the token balances in a bar chart and break it down into the different networks.

def create_bar_chart(api_response):
    # create a list of token balances
    balances = []
    for token in api_response['balances']:
        balances.append(token['balanceUSD'])

    # create a list of token symbols
    symbols = []
    for token in api_response['balances']:
        symbols.append(token['symbol'])

    # create a list of networks
    networks = []
    for token in api_response['balances']:
        networks.append(token['network'])

    # create a bar chart
    import matplotlib.pyplot as plt
    plt.bar(networks, balances)
    plt.show()

create_bar_chart(api_response=response.json())
```

```python
# Now create a chart using matplotlib to show the token balances in a bar chart and break it down into the different networks and tokens.
def create_bar_chart_token_chain(api_response):
    # create a list of token balances
    balances = []
    for token in api_response['balances']:
        balances.append(token['balanceUSD'])

    # create a list of token symbols
    symbols = []
    for token in api_response['balances']:
        symbols.append(token['symbol'])

    # create a list of networks
    networks = []
    for token in api_response['balances']:
        networks.append(token['network'])

    # create a list of tokens
    tokens = []
    for token in api_response['balances']:
        tokens.append(token['name'])

    # create a bar chart
    plt.bar(tokens, balances)
    plt.show()

create_bar_chart_token_chain(api_response=response.json())
```
