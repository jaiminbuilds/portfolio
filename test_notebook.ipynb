```python
wallet = '0x89d9Dd2e85ecC305E276f51BB21fd4C708Be9487'
api = f"https://api.portals.fi/v2/account?owner={wallet}&networks=ethereum&networks=optimism&networks=fantom&networks=arbitrum&networks=polygon&networks=avalanche&networks=bsc"
response = requests.get(api)
print(response.json())
```

```python
# Now create a chart using matplotlib to show the token balances in a pie chart.

def create_chart(api_response):
    # create a list of token balances
    balances = []
    for token in api_response['balances']:
        balances.append(token['balanceUSD'])

    # create a list of token symbols
    symbols = []
    for token in api_response['balances']:
        symbols.append(token['symbol'])

    # create a pie chart
    import matplotlib.pyplot as plt
    plt.pie(balances, labels=symbols)
    plt.show()

create_chart(api_response=response.json())
```
